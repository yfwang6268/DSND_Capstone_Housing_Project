Thus, we will job the columns with more than 40% missing values

we will impute the missing values with the most frequent value of the column for categorical variables and the mean of the column for quantitable variables.

```python
### NA in below columns is not missing data
### it means the property do not have such property
### So we change NA to the new value "NOT_AVALIABLE"

Not_Data_Missing_Cols = ["PoolQC","MasVnrType", "MiscFeature","Alley","Fence", "FireplaceQu","GarageFinish","GarageQual","GarageCond","GarageType",
         "BsmtCond","BsmtExposure","BsmtQual","BsmtFinType2","BsmtFinType1"]
df.loc[:,Not_Data_Missing_Cols].fillna("NOT_AVALIABLE", inplace = True)

# Then we impute the missing values

class DataFrameImputer(TransformerMixin):

    def __init__(self):
        """Impute missing values.

        Columns of dtype object are imputed with the most frequent value 
        in column.

        Columns of other types are imputed with mean of column.

        """
    def fit(self, X, y=None):

        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in X],
            index=X.columns)

        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)

df = DataFrameImputer().fit_transform(df)
df.isnull().sum().sum()
```

##### Transformation

In this part, we perform below transformations and further dummy variables are created based on most of these variables:

* Transform MSSubClass from numerical variables to string variables based on the project documentation, given the reason that MSSubClass represent the type of dwelling and should not be continuous numeric data.
* Transform  below categorical variables to numeric order variables, given the reason that these categorical variables are ordered evaluation and the numeric ordered variables can be feed to the model directly without creating dummy variables.
  + ExterQual ,ExterCond, HeatingQC, GarageQual, GarageCond, KitchenQual
  + BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1, BsmtFinType2

```python
# Transform YrSold and MoSold from numerical variables to cateogorical variables, 
# given the reason that these variables should be categorical data instead of continuous numeric data

df['YrSold'] = df['YrSold'].astype(str)
df['MoSold'] = df['MoSold'].astype(str)

# Transform MSSubClass from numerical variables to string variables based on the project documentation,
# given the reason that MSSubClass represent the type of dwelling and should not be continuous numeric data.

df['MSSubClass'] =  df['MSSubClass'].map({20:'1 story 1946+', 
                           30:'1 story 1945-', 
                           40:'1 story unf attic', 
                           45:'1,5 story unf', 
                           50:'1,5 story fin', 
                           60:'2 story 1946+', 
                           70:'2 story 1945-', 
                           75:'2,5 story all ages', 
                           80:'split/multi level', 
                           85:'split foyer', 
                           90:'duplex all style/age', 
                           120:'1 story PUD 1946+',
                           150:'1,5 story PUD all', 
                           160:'2 story PUD 1946+', 
                           180:'PUD multilevel', 
                           190:'2 family conversion'})

# Transform  below categorical variables to numeric order variables,
# given the reason that these categorical variables are ordered evaluation 
# and the numeric ordered variables can be feed to the model directly without creating dummy variables.

def Score(condition):
    if condition == "NOT_AVALIABLE": result = 0
    elif condition == "GLQ": result = 6
    elif np.isin(condition,["Ex","ALQ"]): result = 5
    elif np.isin(condition,["Gd","BLQ"]): result = 4
    elif np.isin(condition,["Av","Rec","TA"]): result = 3
    elif np.isin(condition,["Fa","LwQ","Mn"]): result = 2
    elif np.isin(condition,["Po","Unf","No"]): result = 1
    else: result = condition
    return result

score_cols_1 = ["ExterQual","ExterCond", "BsmtQual", "BsmtCond", 
                  "BsmtExposure", "HeatingQC", "KitchenQual",
                  "GarageQual", "GarageCond","BsmtFinType1", "BsmtFinType2"]
for col in score_cols_1:
    df[col] = df[col].apply(Score)
```

#### Transfer Year variables to number of years

We would like to transer year variables YearBuilt, YearRemodAdd and GarageYrBlt to number of years by using $2020 - Years$. Since we perform box-cox transformation later, number of years can have a more reasonable value than specific years after box-cox transformation.

```python
year_cols = ["YearBuilt", "YearRemodAdd", "GarageYrBlt"]
for col in year_cols:
    df[col] = 2020 -df[col]

df.loc[df['GarageYrBlt'] < 0, 'GarageYrBlt'] = df.loc[df['GarageYrBlt'] < 0, 'YearBuilt']
```

#### Box-Cox Transformation on Important Quantitative Variables

Below continuous numeric variables are important predictors with right skewness

* LotFrontage, LotArea
* BsmtFinSF1, BsmtFinSF2, BsmtUnfSF, TotalBsmtSF
* X1stFlrSF, X2ndFlrSF,GrLivArea, TotalExtraSF

We apply box-cox transformation on these variables by using power_transform() and specifying method = "box-cox"

```python
X = df.drop(["Id","SalePrice"],axis = 1)
QuantColNames =X.columns.values[X.apply(max,axis = 0) > 10]
X = X[QuantColNames]
X[X<=0] = 0.001
df[QuantColNames] = power_transform(X, method = "box-cox")
```

#### Modelling

We begin this section with $2917$ observations and $212$ predictors. The full list of predictors can be found in the appendix. We split the first $1458$ in the training set and the rest $1459$ into the testing set for Kaggle. Then we split the train set to training set and validation set. Finally, we use the training set to implement Lasso, Elastic Net, Ridge, Gardient Boosting Regression, XGboost regressor and LightGBM regressor to make the predictions. We will use Kaggle score and mean square error as the metrics to evaluate the outcomes.

```python
df.columns = ["".join (c if c.isalnum() else "_" for c in str(x)) for x in df.columns]
# create training set and testing set

training_df = df.iloc[:1458,:]
test_df = df.iloc[1458:,:]
test_Id = test_df.loc[:,'Id']
training_df.drop(['Id'], axis = 1, inplace = True)
test_df.drop(['Id','SalePrice','LogSalePrice'], axis = 1, inplace= True)

# split training set into train set and validation set
y = training_df.LogSalePrice
X = training_df.drop(['SalePrice','LogSalePrice'], axis = 1)
train_X, validation_X, train_y, validation_y = train_test_split(X, y, test_size = 0.1)
```

```python
# below modelling setup is taken from below notebook from kaggle
# https://www.kaggle.com/alexpengxiao/predict-house-prices-model-tuning-ensemble/output
n_folds = 5

def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True).get_n_splits(X)
    rmse = np.sqrt(-cross_val_score(model,
                                    X,
                                    y,
                                    scoring="neg_mean_squared_error",
                                    cv=kf))
    return(rmse)


lasso = make_pipeline(RobustScaler(), Lasso(alpha=0.0005))
ENet = make_pipeline(RobustScaler(), ElasticNet(alpha=0.0005, l1_ratio = 0.9))
ridge = make_pipeline(RobustScaler(), Ridge(alpha=0.0005))
GBoost = GradientBoostingRegressor(n_estimators=3000, learning_rate = 0.05,
                                  max_depth = 4, max_features='sqrt',
                                  min_samples_leaf=15, min_samples_split = 10,
                                  loss= 'huber')
model_xgb = xgb.XGBRegressor(colsample_bytree=0.4603, gamma=0.0468,
                            learning_rate = 0.05, max_depth = 3,
                            min_child_weight = 1.7817, n_estimators=2200,
                            reg_alpha=0.464, reg_lambda = 0.8571,
                            subsample = 0.5213, silent=1,nthread = -1)
model_lgb = lgb.LGBMRegressor(objective = 'regression', num_leaves = 5,
                       learning_rate = 0.05,max_bin = 55,
                       n_estimators = 720,bagging_fraction = 0.8,
                         bagging_freq = 5, feature_fraction = 0.2319,
                         min_data_in_leaf = 6, min_sum_hessian_in_leaf = 11,
                       verbosity = 2)

```

```python
score = rmsle_cv(lasso)
print("\nLasso score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rmsle_cv(ENet)
print("ElasticNet score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rmsle_cv(ridge)
print("Ridge score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rmsle_cv(GBoost)
print("Gradient Boosting score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rmsle_cv(model_xgb)
print("Xgboost score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
score = rmsle_cv(model_lgb)
print("LGBM score: {:.4f} ({:.4f})\n" .format(score.mean(), score.std()))
```

Given aboved XGBoost gives us lowest mean square error, we will tune the XGBoost Model to make the final prediction.

```python
# For Kaggle Score

model_xgb.fit(X,y)
lasso.fit(X,y)
ENet.fit(X,y)
ridge.fit(X,y)
GBoost.fit(X,y)
model_xgb.fit(X,y)
model_lgb.fit(X,y)

pred_lasso = lasso.predict(test_df)
pred_ENet = ENet.predict(test_df)
pred_ridge = ridge.predict(test_df)
pred_GBoost = GBoost.predict(test_df)
pred_xgb = model_xgb.predict(test_df)
pred_lgb = model_lgb.predict(test_df)

# export csv

LASSO_CSV = pd.DataFrame({'Id':test_Id
                      ,"SalePrice":np.exp(pred_lasso)}).to_csv("lasso.csv",index = False)
ENet_CSV = pd.DataFrame({'Id':test_Id
                      ,"SalePrice":np.exp(pred_ENet)}).to_csv("ENet.csv",index = False)
ridge_CSV = pd.DataFrame({'Id':test_Id
                      ,"SalePrice":np.exp(pred_ridge)}).to_csv("ridge.csv",index = False)
GBoost_CSV = pd.DataFrame({'Id':test_Id
                      ,"SalePrice":np.exp(pred_GBoost)}).to_csv("GBoost.csv",index = False)

GBoost_xgb = pd.DataFrame({'Id':test_Id
                      ,"SalePrice":np.exp(pred_xgb)}).to_csv("xgb.csv",index = False)

GBoost_lgb = pd.DataFrame({'Id':test_Id
                      ,"SalePrice":np.exp(pred_lgb)}).to_csv("lgb.csv",index = False)

```

#### Feature Importance

Given the L1 penality on coefficients, there is shrinkage on the coefficients of Lasso model. And for some coefficients $\beta_{j}$, the value shrink to 0, thereby performing variable selection.

Thus, in this section, we extract the non-zero coefficient to check identify which features are importance to predict the housing price.

```python
Scaler = RobustScaler()
Scaled_X = Scaler.fit_transform(X)
lasso = Lasso(alpha=0.0005)
lasso.fit(Scaled_X,y)
```

```python
def feature_importance(classifier, amount, X_train):
    
    feature_importance = abs(classifier.coef_)
    feature_importance = 100.0 * (feature_importance / feature_importance.max())
    fi_len = len(feature_importance)  
    sorted_idx = np.argsort(feature_importance)[range(fi_len-amount, fi_len)]
    pos = np.arange(sorted_idx.shape[0]) + .5
    
    print('Relative Feature Importance Graph')
    featfig = plt.figure(figsize = (8,8))
    featax = featfig.add_subplot(1, 1, 1)
    featax.barh(pos, feature_importance[sorted_idx], align='center')
    featax.set_yticks(pos)
    featax.set_yticklabels(np.array(X_train.columns)[sorted_idx], fontsize=8)
    featax.set_xlabel('Relative Feature Importance')

    plt.tight_layout()   
    plt.show()
    
    ### Code taken from https://stackoverflow.com/questions/24255723/sklearn-logistic-regression-important-features
    

feature_importance(lasso, 20, X)
```
